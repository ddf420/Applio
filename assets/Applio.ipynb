```python
# @title **Install Applio**

import codecs
import time
import os
import csv
import shutil
import tarfile
import subprocess
from pathlib import Path
from datetime import datetime

rot_47_coder = lambda text: "".join(
    [
        (
            chr(
                ((ord(char) - ord("a" if char.islower() else "A") + 47) % 26)
                + ord("a" if char.islower() else "A")
            )
            if char.isalpha()
            else char
        )
        for char in text
    ]
)
rot_47 = lambda encoded_text: "".join(
    [
        (
            chr(
                (ord(c) - (ord("a") if c.islower() else ord("A")) - 47) % 26
                + (ord("a") if c.islower() else ord("A"))
            )
            if c.isalpha()
            else c
        )
        for c in encoded_text
    ]
)

orig_name_of_program = rot_47("Vkkgdj")
new_name_of_program = rot_47("kmjbmvh_hg")
uioawhd = rot_47("cookn://bdocpw.xjh/DVCdnkvij/Vkkgdj.bdo")
uyadwa = codecs.decode("ncc.cl", "rot_13")

from IPython.display import clear_output, Javascript

!git clone --depth 1 $uioawhd
!mv $orig_name_of_program $new_name_of_program
%cd $new_name_of_program/

clear_output()
ForceTemporaryDependence = True
ForceInstallDependencies = True


def vidal_setup(ForceTemporaryStorage, ForceInstallDependencies):

    def acat_install():
        print("Installing pip packages...")
        subprocess.check_call(["pip", "install", "-r", "requirements.txt", "--quiet"])

    if ForceTemporaryStorage:
        file_path = "/content/" + rot_47("Kikpm.ovm.bu")

    content_file_path = "/content/" + rot_47("Kikpm.ovm.bu")
    extract_path = "/"

    if not os.path.exists(file_path):
        folder_path = os.path.dirname(file_path)
        os.makedirs(folder_path, exist_ok=True)
        print("No cached install found..")
        try:
            download_url = rot_47(
                "cookn://cpbbdibavxz.xj/DVCdnkvij/Vkkgdj/mznjgqz/hvdi/Xjgvw/Kikpm.ovm.bu"
            )
            subprocess.run(["wget", "-O", file_path, download_url])
            print("Download completed successfully!")
        except Exception as e:
            print(str(e))

            if os.path.exists(file_path):
                os.remove(file_path)

    if Path(file_path).exists():
        if ForceTemporaryStorage:
            pass
        else:
            shutil.copy(file_path, content_file_path)

        with tarfile.open(content_file_path, "r:gz") as tar:
            for member in tar.getmembers():
                target_path = os.path.join(extract_path, member.name)
                try:
                    tar.extract(member, extract_path)
                except Exception as e:
                    print("Failed to extract a file")
                    ForceInstallDependencies = True
            print(f"Extraction of {content_file_path} to {extract_path} completed.")
        if os.path.exists(file_path):
            os.remove(file_path)

        if ForceInstallDependencies:
            acat_install()
            ForceInstallDependencies = False
    else:
        acat_install()


vidal_setup(ForceTemporaryDependence, ForceInstallDependencies)

clear_output()
print("Finished installing requirements!")
```
